# 파일을 쓰기모드로 열어 출력값 저장하기

```python
f = open("./Data/new.txt","w")

for i in range(1,11):
    data = "%d번째 줄 입니다. \n"%i
    f.write(data)

f.close()
```

```python
# 파일을 읽기모드로 열어 파일 내용 출력하기 - 1

f = open("./Data/new.txt","r")
while True:
    line = f.readline()# f.readlines 줄 전체 
    if not line:
        break
    print(line)
f.close()
```

```python
# 파일을 읽기모드로 열어 파일 내용 출력하기 - 2
# 데이터가 무지 많을때에는 이쪽보다는 위에꺼 방향으로 
f = open("./Data/new.txt", "r")
lines = f.readlines()
for line in lines:
    print(line)
f.close()
```

```python
# 파일을 읽기모드로 열어 파일 내용 출력하기 - 3
f = open("./Data/new.txt","r")
data = f.read()
print(data)
f.close
```

```python
# 파일에 새로운 내용 추가하기

f = open('./Data/new.txt','a')
for i in range(11,21):
    data = "%d번째 추가한 줄 입니다.\n"%i
    f.write(data)

f.close()
```

```python
f = open("./Data/new.txt","r")
data = f.read()
print(data)
f.close
```

### with

```python
# make file
with open('./Data/new2.txt',"w") as f:
    for i in range(1,11):
        data = "%d번째 줄 입니다.\n"%i
        f.write(data)
```

```python
# read file

with open('./Data/new2.txt',"r") as f:
    data=f.read()
    print(data)
```
